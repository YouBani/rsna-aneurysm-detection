## Check Subtype

```python
from collections import Counter
from pathlib import Path
import pydicom

data_root = Path('/fsx')
series_dir = data_root / "raw" / "series"

series_descriptions = []
char_sets = []
procedure_code_meanings = []

for uid_dir in series_dir.iterdir():
    files = sorted(uid_dir.glob("*.dcm"))
    if files:
        ds = pydicom.dcmread(str(files[0]))
        
        # 1. Collect SeriesDescription
        desc = str(getattr(ds, "SeriesDescription", "")).upper()
        if desc:
            series_descriptions.append(desc)

        # 2. Collect Specific Character Set
        char_set = str(getattr(ds, "SpecificCharacterSet", "")).upper()
        if char_set:
            char_sets.append(char_set)

        # 3. Collect Procedure Code Meanings
        try:
            procedure_codes = getattr(ds, "ProcedureCodeSequence", [])
            for item in procedure_codes:
                code_meaning = str(getattr(item, "CodeMeaning", "")).upper()
                if code_meaning:
                    procedure_code_meanings.append(code_meaning)
        except Exception:
            pass
    
print("--- DICOM Metadata Summary ---")
print("\nSeries Descriptions:")
print(Counter(series_descriptions))

print("\nSpecific Character Sets:")
print(Counter(char_sets))

print("\nProcedure Code Meanings:")
print(Counter(procedure_code_meanings))
print("-----------------------------")     

    
```

```python
import os
import argparse
from pathlib import Path
from collections import Counter
from pydicom import dcmread

def check_subtypes(data_root: str):
    """
    Scans the dataset for specific DICOM metadata fields and prints a summary
    of their values. This helps in understanding data characteristics and
    potential inconsistencies like different languages.
    """
    data_root = Path(data_root)
    series_dir = data_root / "raw" / "series"
    
    if not series_dir.exists():
        raise FileNotFoundError(f"Series directory not found at: {series_dir}")

    series_descriptions = []
    char_sets = []
    procedure_code_meanings = []
    
    for uid_dir in series_dir.iterdir():
        if uid_dir.is_dir():
            # Get the first DICOM file from the directory to check metadata
            files = sorted(uid_dir.glob("*.dcm"))
            if files:
                ds = dcmread(str(files[0]))

                # 1. Collect SeriesDescription
                desc = str(getattr(ds, "SeriesDescription", "")).upper()
                if desc:
                    series_descriptions.append(desc)

                # 2. Collect Specific Character Set
                char_set = str(getattr(ds, "SpecificCharacterSet", "")).upper()
                if char_set:
                    char_sets.append(char_set)

                # 3. Collect Procedure Code Meanings
                try:
                    procedure_codes = getattr(ds, "ProcedureCodeSequence", [])
                    for item in procedure_codes:
                        code_meaning = str(getattr(item, "CodeMeaning", "")).upper()
                        if code_meaning:
                            procedure_code_meanings.append(code_meaning)
                except Exception:
                    pass
    
    print("--- DICOM Metadata Summary ---")
    print("\nSeries Descriptions:")
    print(Counter(series_descriptions).most_common(10))
    
    print("\nSpecific Character Sets:")
    print(Counter(char_sets).most_common(10))
    
    print("\nProcedure Code Meanings:")
    print(Counter(procedure_code_meanings).most_common(10))
    print("-----------------------------")


if __name__ == "__main__":
    ap = argparse.ArgumentParser(description="Check DICOM subtypes in the dataset.")
    ap.add_argument("--data_root", required=True, help="Path to the dataset's root directory.")
    args = ap.parse_args()

    check_subtypes(args.data_root)
```
